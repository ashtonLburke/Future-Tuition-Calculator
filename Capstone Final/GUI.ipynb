# Welcome to the Tuiton Calculator!

## Tuition costs for undergraduate universities has been steadily rising over the years. This can make it difficult to estimate how much money you will need to enroll. 

```python
from sklearn.linear_model import Ridge
df = pd.read_csv('AVG_UNI_COST.csv')
lr = Ridge()
fig = plt.figure(figsize = (10, 5))
lr.fit(df[['Year']], df['Value'])
plt.bar(df['Year'], df['Value'])
plt.plot(df['Year'], lr.coef_*df['Year']+lr.intercept_, color='orange')
plt.xlabel("Year of Admission")
plt.ylabel("Cost of Tuition")
plt.title("Cost of Tuition Over the Last Fifty Years")
plt.show()
```

## This tuition calculator was designed to help address this problem. Simply input the year you want to enroll and the calculator will give you an estimate on tuition costs for that year.

### Please enter your intended year of enrollment below!

```python
import sys
import time

def delete_last_line():
    "Use this function to delete the last line in the STDOUT"

    #cursor up one line
    sys.stdout.write('\x1b[1A')

    #delete last line
    sys.stdout.write('\x1b[2K')



text = widgets.Text(
    value='Enter here!',
    placeholder='Type something',
    description='',
    disabled=False
)
display(text)

def callback(wdgt):
    if wdgt.value.isdigit():
        y = 275 * int(wdgt.value) + -539662
    
        print ("You entered the year: ", wdgt.value, ". Therefore, your expected tution cost is: $", y)

    else:
        print ("Please enter a valid year")
        time.sleep(2)
        delete_last_line()


text.on_submit(callback)
```

### This calculator utilizes a linear regression model for it's predictions. These predictions should be taken as facts but rather rough estimates.

```python
from sklearn.linear_model import Ridge
db = pd.read_csv('AVG_UNI_COST.csv')
lr = Ridge()
fig = plt.figure(figsize = (10, 5))
lr.fit(db[['Year']], db['y_pred'])
plt.bar(db['Year'], db['y_pred'])
plt.plot(db['Year'], lr.coef_*db['Year']+lr.intercept_, color='orange')
plt.xlabel("Year of Admission")
plt.ylabel("Model's Predicted Cost of Tuition")
plt.title("Predicted Cost of Tuition Over the Last Fifty Years")
plt.show()
```

# As seen above, the linear regression model utilizes a trendline to make predictions. The trendline was generated from historical tuition data. 

### Mapping out both the actual data and predicted data together will allow us to visualize the accuracy of this model

### Please double-click the graph to zoom in

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('AVG_UNI_COST.csv')
labels = df["Year"]
historical = df["Value"]
predicted = df["y_pred"]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(30,15))
rects1 = ax.bar(x - width, historical, width, label='Historical Tuition Data')
rects2 = ax.bar(x + width, predicted, width, label='Predicted Tuition Data')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Tuition Cost',fontsize = 40)
ax.set_xlabel('Year of Enrollment', fontsize = 40)
ax.set_title('Historical Tuition Cost Versus Predicted Tuition Cost', fontsize = 40)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize = 12)
ax.legend(fontsize = 40)




fig.tight_layout()

plt.show()
```
